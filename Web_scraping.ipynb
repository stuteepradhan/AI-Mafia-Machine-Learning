```python
import pandas as pd
from pandas import Series,DataFrame
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.studentnewsdaily.com/'
result=requests.get(url)
print(result.content)
```

```python
soup=BeautifulSoup(result.content,'lxml')
```

```python
print(soup)
```

```python
content=soup.find_all("div",class_='list-group list-group-flush')
for i in content:
    print(i.get_text())
```
