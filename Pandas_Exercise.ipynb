## Reading data from a CSV file

You can read data from a CSV file using the ``read_csv`` function. By default, it assumes that the fields are comma-separated.

````python
import pandas as pd# import pandas
````

>The `imdb.csv` dataset contains Highest Rated IMDb "Top 1000" Titles.

````python
imdb_df=pd.DataFrame(pd.read_csv('imdb_1000.csv'))
imdb_df# load imdb dataset as pandas dataframe
````

````python
imdb_df.head()# show first 5 rows of imdb_df
````

>The `bikes.csv` dataset contains information about the number of bicycles that used certain bicycle lanes in Montreal in the year 2012.

````python
bikes_df=pd.DataFrame(pd.read_csv('bikes.csv',sep=';',parse_dates=['Date']))
bikes_df# load bikes dataset as pandas dataframe
````

````python
bikes_df.head(3)# show first 3 rows of bikes_df
````

## Selecting columns

When you read a CSV, you get a kind of object called a DataFrame, which is made up of rows and columns. You get columns out of a DataFrame the same way you get elements out of a dictionary.

````python
imdb_col=imdb_df.keys()
print(imdb_col)# list columns of imdb_df
````

````python
imdb_df.dtypes# what are the datatypes of values in columns
````

````python
print(imdb_df['title'][:5])# list first 5 movie titles
````

````python
imdb_df[['title','genre']]# show only movie title and genre
````

## Understanding columns

On the inside, the type of a column is ``pd.Series`` and pandas Series are internally numpy arrays. If you add ``.values`` to the end of any Series, you'll get its internal **numpy array**.

````python
print(imdb_df['duration'].dtypes)
print(type(imdb_df['duration']))# show the type of duration column
````

````python
import numpy as np
my_arr=np.array(imdb_df['duration'])
print(my_arr)# show duration values of movies as numpy arrays
````

## Applying functions to columns

Use `.apply` function to apply any function to each element of a column.

````python
imdb_df['title'].apply(lambda x:x.upper())# convert all the movie titles to uppercase
````

## Value counts

Get count of unique values in a particular column/Series.

````python
imdb_df.genre.unique()# what are the unique genre in imdb_df?
````

````python
imdb_df['genre'].value_counts().plot(kind='bar',title='Genres')# plotting value counts of unique genres as a bar chart
````

````python
imdb_df['genre'].value_counts().plot(autopct = '%1.1f%%',kind='pie',radius=1,title='Genre',figsize=(40,10))# plotting value counts of unique genres as a pie chart
````

## Group By

Any groupby operation involves one of the following operations on the original object. They are −

- Splitting the Object

- Applying a function

- Combining the results

In many situations, we split the data into sets and we apply some functionality on each subset. In the apply functionality, we can perform the following operations −

- **Aggregation** − computing a summary statistic

- **Transformation** − perform some group-specific operation

- **Filtration** − discarding the data with some condition

````python

imdb_df.groupby(['genre'])#group imdb_df by movie genres
````

````python
x=imdb_df.groupby(imdb_df['genre']).get_group('Crime')
x# get crime movies group
````

````python
x=imdb_df.groupby(['genre']).mean()
x['duration']

    # get mean of movie durations for each group
````

````python

my_crime=imdb_df.groupby('genre').get_group('Crime')
my_crime['duration']=my_crime['duration'].mean()
my_crime[['title','duration']]



# change duration of all movies in a particular genre to mean duration of the group
````

````python
imdb_df.groupby('genre').filter(lambda a:a['duration'].mean()>float(120))


# drop groups/genres that do not have average movie duration greater than 120.
````

````python
x=bikes_df.groupby('weekday').sum()
# group weekday wise bikers count
````

````python
x# get weekday wise biker count
````

````python
plt.plot(x['Berri1'],color='red')# plot weekday wise biker count for 'Berri1'
````
