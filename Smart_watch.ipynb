```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#data loading
xtrain=pd.read_csv('Linear_X_Train.csv')
ytrain=pd.read_csv('Linear_Y_Train.csv')
xtest=pd.read_csv('Linear_X_Test.csv')
```

```python
xtrain=xtrain.values
ytrain=ytrain.values
xtest=xtest.values
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
#Training the model
model.fit(xtrain,ytrain)
```

```python
#Testing model
my_prediction=model.predict(xtest)
print('The performance prediction is:')
for i in range(len(my_prediction)):
    print(my_prediction[i])
```

```python
#Prediction
plt.scatter(xtrain,ytrain,label='Training set')
plt.plot(xtest,my_prediction,label='Prediction of performance',color='red')
plt.title('This model predicts level of performance students achieved by devoting the given time.')
plt.legend()
plt.show()
```

```python
#Accuracy of the model
print('The accuracy of the model is:')
model.score(xtrain,ytrain)
```
